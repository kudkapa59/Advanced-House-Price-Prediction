## Missing values

### Categorical

```python
categorical = [features for features in dataset.columns if dataset[features].dtypes == 'O' and dataset[features].isnull().sum()>=1]
for i in categorical:
    print(i,"Percentage of missing values","{:.2%}".format(dataset[i].isnull().sum()/dataset.shape[0]))
```

```python
for i in categorical:
    #data = dataset.copy()
    dataset[i] = dataset[i].fillna('Missing')
for i in categorical:
    print(i,"Percentage of missing values","{:.2%}".format(dataset[i].isnull().sum()/dataset.shape[0]))
```

```python
dataset.head()
```

### Numerical

```python
numerical = [features for features in dataset.columns if dataset[features].dtypes != 'O' and dataset[features].isnull().sum()>1]
for i in numerical:
    print(i,"Percentage of missing values","{:.2%}".format(dataset[i].isnull().sum()/dataset.shape[0]))
```

```python
for i in numerical:
    #data = dataset.copy()
    m = dataset[i].median()
    dataset[i+'NAN'] = np.where(dataset[i].isnull(),1,0)
    dataset[i].fillna(m, inplace = True)
for i in numerical:
    print(i,"Percentage of missing values","{:.2%}".format(dataset[i].isnull().sum()/dataset.shape[0]))
```

```python
dataset.head()
```

```python
year_variables = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']
for i in year_variables:
    dataset[i] = dataset['YrSold'] - dataset[i]
```

```python
dataset.head()
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

### Continuous Features

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

## Rare Categorical Variables

We are going to remove the categorical variables, that contribute to only 1% of the observations

```python
#extracting the categorical features
categorical_features = [features for features in dataset.columns if dataset[features].dtype == 'O']
len(categorical_features)
```

```python
#Replace rare variables with a string "Rare Variable"
for feature in categorical_features:
    p = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    p = p[p>0.01].index
    dataset[feature] = np.where(dataset[feature].isin(p),dataset[feature],'Rare Variable')
```

```python
#Replacing categorical variables with the rank of popularity in the observations
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head()
```

## Feature Scaling

```python
#Features, which we want to scale
scale_features = [features for features in dataset.columns if features not in ['Id','SalePrice']]
len(scale_features)
```

```python
#Normalizing the data
scaler = MinMaxScaler()
scaler.fit(dataset[scale_features])
```

```python
#Concatenation of the scaled and not-scaled features
dataset = pd.concat([dataset[['Id','SalePrice']].reset_index(drop = True),
                     pd.DataFrame(scaler.transform(dataset[scale_features]),columns = scale_features)],axis = 1)
```
